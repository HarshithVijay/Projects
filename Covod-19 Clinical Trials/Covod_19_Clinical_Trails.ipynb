```python
from sklearn.model_selection import train_test_split
```

```python
#converting string date to date-time
import datetime as dt
india_case['date'] = pd.to_datetime(india_case['date'])
india_case.head()
```

```python
#conerting date-time to ordinal
india_case['date']=india_case['date'].map(dt.datetime.toordinal)
india_case.head()
```

```python
india_case.tail()
```

```python
#getting dependent variable and independent
x = india_case['date'] #independent variable that i use to predict the dependent variable
y = india_case['total_cases'] # dependent variable that i want to predict
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3)
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
import numpy as np
lr.fit(np.array(x_train).reshape(-1,1),np.array( y_train).reshape(-1,1))
```

```python
y_pred=lr.predict(np.array(x_test).reshape(-1,1))
```

```python
#	738586 = 2023-03-07,,,,so i'll predict the cases to 2023-03-10 i.e 738589
lr.predict(np.array([[738589]]))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
#	738586 = 2023-03-07,,,,so i'll predict the cases to 2023-03-10 i.e 738589
rf.predict([[738589]])
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
#	738586 = 2023-03-07,,,,so i'll predict the cases to 2023-03-10 i.e 738589
dtr.predict([[738589]])
```
